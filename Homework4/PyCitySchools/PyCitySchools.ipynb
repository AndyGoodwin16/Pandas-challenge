```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "../Resources/schools_complete.csv"
student_data_to_load = "../Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculating stats for entire district.
total_schools = school_data['school_name'].nunique()

total_students = student_data['student_name'].count()

total_budget = school_data['budget'].sum()

mean_math = student_data['math_score'].mean()

mean_reading = student_data['reading_score'].mean()

math_pass = student_data[student_data['math_score'] >= 70]
perc_math = math_pass['student_name'].count()/student_data['student_name'].count()*100

reading_pass = student_data[student_data['reading_score'] >= 70]
perc_reading = reading_pass['student_name'].count()/student_data['student_name'].count()*100

overall_pass = student_data[(student_data['math_score'] >= 70) & (student_data['reading_score'] >= 70)]
perc_overall = overall_pass['student_name'].count()/student_data['student_name'].count()*100
```

```python
#Dataframe and formating for district stats.
district_summary_df = pd.DataFrame({
    "Total Schools": total_schools,
    "Total Students": f"{total_students:,}",
    "Total Budget": f"${total_budget:,.2f}",
    "Average Math Score": f"{mean_math:.2f}",
    "Average Reading Score": f"{mean_reading:.2f}",
    "% Passing Math": f"{perc_math:.2f}%",
    "% Passing Reading": f"{perc_reading:.2f}%",
    "% Overall Passing": f"{perc_overall:.2f}%"
}, index=[0])

district_summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#Calculating stats broken down by school. Generating lists for relevant categories. For loops to make lists from math operations on two existing lists.
school_data = school_data.sort_values('school_name')

school_names = school_data['school_name'].tolist()

school_types = school_data['type'].tolist()

school_total_students = school_data['size'].tolist()

school_total_budget = school_data['budget'].tolist()

per_student_budget = []
for i in range(len(school_total_students)):
    per_student_budget.append(school_total_budget[i]/school_total_students[i])

group_school_data_complete = school_data_complete.set_index('school_name').groupby(['school_name'])

school_mean_math = group_school_data_complete['math_score'].mean().tolist()

school_mean_reading = group_school_data_complete['reading_score'].mean().tolist()

school_math_pass = school_data_complete[school_data_complete['math_score'] >= 70].groupby('school_name')['Student ID'].count().tolist()
school_math_perc = []
for i in range(len(school_math_pass)):
    school_math_perc.append(school_math_pass[i]/school_total_students[i]*100)

school_reading_pass = school_data_complete[school_data_complete['reading_score'] >= 70].groupby('school_name')['Student ID'].count().tolist()
school_reading_perc = []
for i in range(len(school_reading_pass)):
    school_reading_perc.append(school_reading_pass[i]/school_total_students[i]*100)

school_overall_pass = school_data_complete[(school_data_complete['math_score'] >= 70) & (school_data_complete['reading_score'] >= 70)].groupby('school_name')['Student ID'].count().tolist()
school_overall_perc = []
for i in range(len(school_overall_pass)):
    school_overall_perc.append(school_overall_pass[i]/school_total_students[i]*100)
```

```python
#Dataframe and formating for school stats.
school_summary_df = pd.DataFrame({
    "": school_names,
    "School Type": school_types,
    "Total Students": school_total_students,
    "Total School Budget": school_total_budget,
    "Per Student Budget": per_student_budget,
    "Average Math Score": school_mean_math,
    "Average Reading Score": school_mean_reading,
    "% Passing Math": school_math_perc,
    "% Passing Reading": school_reading_perc,
    "% Overall Passing": school_overall_perc
})
school_summary_df = school_summary_df.set_index([""])
school_summary_df.style.format({
    "Total Students": '{:}',
    "Total School Budget": "${:,.2f}",
    "Per Student Budget": "${:.2f}",
    'Average Math Score': "{:.2f}", 
    'Average Reading Score': "{:.2f}", 
    "% Passing Math": "{:.2f}%", 
    "% Passing Reading": "{:.2f}%",
    "% Overall Passing": "{:.2f}%"
})
```

## Top Performing Schools (By % Overall Passing)

```python
#Sorting schools by descending Overall Passing %.
top_schools_df = school_summary_df.sort_values("% Overall Passing", ascending=False).head()
#Dataframe and formating for school stats.
top_schools_df.style.format({
    "Total Students": '{:}',
    "Total School Budget": "${:,.2f}",
    "Per Student Budget": "${:.2f}",
    'Average Math Score': "{:.2f}", 
    'Average Reading Score': "{:.2f}", 
    "% Passing Math": "{:.2f}%", 
    "% Passing Reading": "{:.2f}%",
    "% Overall Passing": "{:.2f}%"
})
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#Creating bins for ranges of school spending per student. Grouping data and assigning each student to a bin.
spending_bins = [578, 597, 617, 637, 655]
spending_groups = ["$578-597", "$598-617", "$618-637", "$638-655"]
school_data_complete['spending_range'] = pd.cut(school_data_complete['budget']/school_data_complete['size'], spending_bins, labels=spending_groups)
spending = school_data_complete.groupby('spending_range')
#Calculating stats for assigned bins.
spending_math_mean = spending['math_score'].mean()
spending_reading_mean = spending['reading_score'].mean()
spending_math_perc = school_data_complete[school_data_complete['math_score'] >= 70].groupby('spending_range')['Student ID'].count()/spending['Student ID'].count()*100
spending_reading_perc = school_data_complete[school_data_complete['reading_score'] >= 70].groupby('spending_range')['Student ID'].count()/spending['Student ID'].count()*100
spending_overall_perc = school_data_complete[(school_data_complete['math_score'] >= 70) & (school_data_complete['reading_score'] >= 70)].groupby('spending_range')['Student ID'].count()/spending['Student ID'].count()*100
```

```python
#Dataframe and formating for bins of school spending per student.
spending_df = pd.DataFrame({
    "Average Math Score": spending_math_mean,
    "Average Reading Score": spending_reading_mean,
    "% Passing Math": spending_math_perc,
    "% Passing Reading": spending_reading_perc,
    "% Overall Passing": spending_overall_perc
})
spending_df.index.name = 'Spending Ranges (Per Student)'
spending_df.style.format({
    "Average Math Score": "{:.2f}",
    "Average Reading Score": "{:.2f}",
    "% Passing Math": "{:.2f}%",
    "% Passing Reading": "{:.2f}%",
    "% Overall Passing": "{:.2f}%"
})
```
